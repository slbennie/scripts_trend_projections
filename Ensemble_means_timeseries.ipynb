```python
#basically looping through all the ensemble means to get their timeseriese data
#also noting the names of each model in model_names (not doing the ensemble number yet)
ds_era5  = xr.open_dataset('../../../../../gws/nopw/j04/extant/users/slbennie/CVDP_outputs/NAO/historical_NAO/ERA5_1.cvdp_data.1940-2014.nc')
nao_timeseries_djf_era5 = ds_era5['nao_timeseries_djf']
time_era5 = np.arange(1940,2015)

model_names = []
time = np.arange(1850,2015)
fig, ax = plt.subplots(figsize=(20,10))
from pathlib import Path

count = 0

# Directory where your .nc files are located
directory = Path('../../../../../gws/nopw/j04/extant/users/slbennie/CVDP_outputs/NAO/historical_NAO/')

# Loop through all .nc files in the directory
for file_path in directory.iterdir():
    if file_path.is_file() and file_path.suffix == '.nc':
        if '_EM.cvdp' in str(file_path):
            model_names.append(file_path.name.split('_')[0])
            ds = xr.open_dataset(file_path)
            nao_timeseries_djf_em = ds['nao_timeseries_djf_em']
            if model_names[count] == 'ACCESS-ESM1-5':
                ax.plot(time,nao_timeseries_djf_em, label=model_names[count], color='#1f77b4')
            if model_names[count] == 'CanESM5':
                ax.plot(time,nao_timeseries_djf_em, label=model_names[count], color='#ff7f0e')
            if model_names[count] == 'CMCC-CM2-SR5':
                ax.plot(time,nao_timeseries_djf_em, label=model_names[count], color='#FFD700')
            if model_names[count] == 'FGOALS-g3':
                ax.plot(time,nao_timeseries_djf_em, label=model_names[count], color='#2ca02c')
            if model_names[count] == 'GISS-E2-1-G':
                ax.plot(time,nao_timeseries_djf_em, label=model_names[count], color='#d62728')
            if model_names[count] == 'HadGEM3-GC31-LL':
                ax.plot(time,nao_timeseries_djf_em, label=model_names[count], color='#9467bd')
            if model_names[count] == 'IPSL-CM6A-LR':
                ax.plot(time,nao_timeseries_djf_em, label=model_names[count], color='#8c564b')
            if model_names[count] == 'MIROC6':
                ax.plot(time,nao_timeseries_djf_em, label=model_names[count], color='#e377c2')
            if model_names[count] == 'MPI-ESM1-2-LR':
                ax.plot(time,nao_timeseries_djf_em, label=model_names[count], color='#7f7f7f')
            if model_names[count] == 'NorESM2-LM':
                ax.plot(time,nao_timeseries_djf_em, label=model_names[count], color='#bcbd22')
            
            count = count + 1

print(np.shape(nao_timeseries_djf_era5), len(time))
ax.plot(time_era5,nao_timeseries_djf_era5.values, label='ERA5', color='black', linewidth=3)
ax.axhline(0, color='black')
ax.legend(ncols=3, fontsize = 'xx-small')
ax.set_title('NAO Ensemble means - Historical')
ax.set_xlabel('Time (years)')
ax.set_ylabel('NAO Index')
#ax.set_xlim(1940,2015)

plt.savefig('NAO_plots/NAO_historical_EM.png')
#its gonna be funny if ending in 2014 instead of 2020 for some files!!!
print(model_names)
```
