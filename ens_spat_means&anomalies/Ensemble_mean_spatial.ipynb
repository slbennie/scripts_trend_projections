```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import os
import cftime
import pandas as pd

def get_files_for_model_exp(folder_path, experiment):
    model_exp_files = [os.path.join(folder_path, filename) for filename in os.listdir(folder_path) if experiment in filename]
    return model_exp_files

def calculate_mean_between_ensembles(file_paths, output_path, variable):
    #wanting to calculate the ensemble mean across multiple files with the structure time,lat,lon (2052,71,144)
    #going to open the files and then calculate the mean across the ensembles
    #will then save to an output file given by output_path
    #file_paths is the list of files - hoping there is a wildcard that can be used when calling the function.
    #variable is the variable from the file - need to be given as a string 'psl', 'tos', etc.

    #Open the files
    ds = xr.open_mfdataset(file_paths, combine='nested', concat_dim='ensemble')

    #calculate the mean
    mean = ds[variable].mean(dim='ensemble')

    #save results to a netcdf file
    mean.to_netcdf(output_path)

    ds.close()

    print(f"Mean data saved to {output_path}")
```

```python
def calc_seasonal_ens_mean_spat(filepath, season, variable, output_path):
    #Passing through the filepath for the file that has the ensemble mean spatial pattern timeseries
    #season = string 'djf', etc. -
    #variable - 'psl', 'tos', etc.
    #output_path - where to send the .nc file of the seasonal ensemble mean spatial pattern timeseries

    #opening the ensemble mean spatial pattern file
    file = xr.open_dataset(filepath)

    #initialising lists to hold the seasonal mean (one mean spatial pattern per year for that season, will only keep the djf mean)
    #also creating a list for the new times - need to still have the years associated going from 12 values a year to 1.
    seasonal_mean_list = []
    new_times = []

    #defining the starting index for each season depending on which season is given
    #this will start from the first ear of the timeseries - to change this would need to pass a start year through, put in index form and then add to below
    if season == 'mam':
        start = 2
    elif season == 'jja':
        start = 5
    elif season == 'son':
        start = 8
    elif season == 'djf':
        start = 11
    else:
        start = len(file['time'])
        print('This is not a standard season option')

    #this will go to the correct month, end_of_data, skips to the next year (start,len(file['time']),12)
    for i in range(start,len(file['time']),12):
        #print('The index', i)

        #takes a slice of the season
        time_slice = file[variable].isel(time=slice(i,i+3)).values
    
        # Compute the mean across the time axis (axis=0 corresponds to the time dimension)
        seasonal_mean = np.mean(time_slice, axis=0)
        #print(seasonal_mean)
    
        #append seasonal mean to corresponding sesonal list 
        seasonal_mean_list.append(seasonal_mean)

        #appending the times that will match the first month of the season so djf will be d's year
        new_times.append(file['time'][i].values)
    
    #convert list into an array
    seasonal_mean_array = np.array(seasonal_mean_list)
    #print(type(new_times),new_times)
    new_times_converted = [pd.to_datetime(str(time)) for time in new_times]
    new_times_converted = pd.to_datetime(new_times_converted)
    #print(type(new_times_converted),new_times_converted)
    
    #create the lat and lon (extract from the previous files grid)
    latitudes = file['lat'].values
    longitudes = file['lon'].values

    #create a dataset of the seasonal means to then create a .nc file
    #want to index by the lat and lon from the non-seasonal spatial pattern ens mean
    #and index by the year where the year is that which is in the seasons first month
    seasonal_mean_dataset = xr.Dataset(
        {
            variable: (['time', 'lat', 'lon'], seasonal_mean_array)
        },
        coords={
            'time': new_times_converted,  # Time axis
            'lat': latitudes,
            'lon': longitudes
        }
    )
    encoding = {
        'time': {'units': 'days since 0001-01-01 00:00:00', 'calendar': '360_day'}
    }
    
    seasonal_mean_dataset.to_netcdf(output_path, encoding=encoding)
    
    print("NetCDF file with mean values has been saved!")
```

```python
season = 'son'
variable = 'psl'
frequency = 'mon'
experiment = 'historical'
model = 'HadGEM3-GC31-LL'
filepath_ens_mean_spat = '/../../gws/nopw/j04/extant/users/slbennie/CVDP_outputs/ens_mean_spat/psl/ens_mean_spat'+'_'+variable+'_'+frequency+'_'+experiment+'_'+model+'.nc'
output_path = '/gws/nopw/j04/extant/users/slbennie/CVDP_outputs/ens_mean_spat/psl/seasonal_means/' + 'ens_mean_' + season + '_' + variable + '_' + frequency + '_' + experiment + '_' + model + '.nc'

calc_seasonal_ens_mean_spat(filepath_ens_mean_spat, season, variable, output_path)
```
