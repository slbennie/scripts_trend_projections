```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
import cartopy.feature as cfeature
from eofs.xarray import Eof
import cartopy.crs as ccrs
import cftime
import os
```

```python
#FUNCTIONS TO CROP BOTH THE HISTORICAL EM FROM CVDP AND THE PSL ANOMALIES TO THE NA SECTOR
#ALSO ENSURES THAT THE COORDINATE SYSTEMS LINE UP

#CVDP'S HISTORICAL EM
def CVDP_EM_crop_NA_sector(filename, pattern):
    #function which will crop the historical ensemble mean CVDP output to the NA sector
    ds = xr.open_dataset(filename)
    ds = ds[pattern]
    
    #finding the longitudes that are greater than 179
    new_lon = np.where(ds.lon > 179, ds.lon -360, ds.lon)
    
    #creating a copy of the data array where the longitudes have been shifted
    ds_shifted = ds.copy()

    #adding the new longitudes to ds_shifted
    ds_shifted.coords['lon'] = new_lon
    
    #Now need to make sure they are in the correct order and then re-index to make sure the lon get put to match the sorted lon
    sorted_lon = np.sort(ds_shifted.lon)
    ds_shifted = ds_shifted.sel(lon=sorted_lon)
    
    historical_NAO_EM_shifted = ds_shifted.sel(lat=slice(20,80), lon=slice(-90,40))

    return historical_NAO_EM_shifted

#PSL ANOMALIES
def crop_ens_north_atlantic(filename, year_initial, year_final):
    #function to crop an ensemble member to the north atlantic region and time
    anomaly = xr.open_dataset(filename)
    anomaly_time = anomaly.sel(year=slice(year_initial, year_final))
    anomaly_NA_time = anomaly_time.sel(lat=slice(20,80), lon=slice(-90,40))

    return anomaly_NA_time
```

```python
#FUNCTION - project_onto_spatial_pattern
#this will project an anomaly onto a spatial pattern

#INPUTs
#anomaly - xarray dataset from an opened a .nc file, contains a lat lon time grid for psl, cut to NA
#spat_pattern - xarray dataset from an opened a .nc file containing a lat lon grid for psl values, cut to NA and shifted so coordinate systems match
#             - already selected the pattern 'nao_pattern_djf_em'
#variable - for example 'psl'

#OUTPUTS
#NAO_index - will also save this as a .nc?

def project_onto_spatial_pattern(anomaly, spat_pattern, variable, output_file):
    #function which will project an anomaly (lat,lon,time) onto a spatial pattern (lat,lon)

    # Weight psl data by coslat to account for grid cell area decreasing with latitude
    weights = np.cos(np.radians(anomaly["lat"].values))
    weights_2d = weights[:, np.newaxis]

    # weight psl (or another variable) anomalies by area of each gridcell
    weighted_var = anomaly[variable] * weights_2d

    # also weight the spatial pattern and *100 to convert from hPa into Pa to match the anomalies
    weighted_regression = np.nan_to_num((spat_pattern*100) * weights_2d)

    # Combine lat and lon dimensions
    weighted_var_flat = weighted_var.stack(spatial=['lat','lon'])

    # flatten weighted regression pattern
    regression_flat = weighted_regression.flatten()

    # dot product (weighted projection) between the weighted psl anomalies and regression pattern
    weighted_projection = (weighted_var_flat * regression_flat).sum(dim="spatial")

    # normalise for unit variance
    total_variance = np.sum(regression_flat**2)

    NAO_index = weighted_projection / total_variance

    #save to a netCDF file
    NAO_index.to_netcdf(output_file)
    return NAO_index
```

```python
#function which defines the models used for each experiment.
def get_models_for_experiment(experiment):
    if experiment == 'historical':
        model = ['ACCESS-ESM1-5','CanESM5','CMCC-CM2-SR5','FGOALS-g3','GISS-E2-1-G','HadGEM3-GC31-LL','IPSL-CM6A-LR','MIROC6','MPI-ESM1-2-LR','NorESM2-LM']
    elif experiment == 'hist-aer':
        model = ['ACCESS-ESM1-5','CanESM5','CMCC-CM2-SR5','FGOALS-g3','GISS-E2-1-G','HadGEM3-GC31-LL','IPSL-CM6A-LR','MIROC6','MPI-ESM1-2-LR','NorESM2-LM']#'CNRM-CM6-1' to be added once CVDP EM calc
    elif experiment == 'hist-GHG':
        model = ['ACCESS-ESM1-5','CanESM5','CMCC-CM2-SR5','FGOALS-g3','GISS-E2-1-G','HadGEM3-GC31-LL','IPSL-CM6A-LR','MIROC6','MPI-ESM1-2-LR','NorESM2-LM']#'CNRM-CM6-1' to be added once CVDP EM calc
    elif experiment == 'hist-sol':
        model = ['ACCESS-ESM1-5','CanESM5','GISS-E2-1-G','HadGEM3-GC31-LL','MIROC6','MPI-ESM1-2-LR','NorESM2-LM']
    elif experiment == 'hist-totalO3':
        model = ['CanESM5','GISS-E2-1-G','HadGEM3-GC31-LL','MIROC6','MPI-ESM1-2-LR','NorESM2-LM']
    elif experiment == 'hist-volc':
        model = ['ACCESS-ESM1-5','CanESM5','CMCC-CM2-SR5','GISS-E2-1-G','HadGEM3-GC31-LL','MIROC6','MPI-ESM1-2-LR','NorESM2-LM']

    return model
```

```python
#setting up variables
variable = 'psl'

#key for CVDP's historical ensemble mean pattern which then ensemble members will be projected onto
pattern = 'nao_pattern_djf_em'

#list of experiments
experiment = ['historical', 'hist-aer', 'hist-GHG', 'hist-sol', 'hist-totalO3', 'hist-volc']

for e in experiment:
    #looping through each experiment
    print("Calculating the projection of the djf psl anomaly onto CVDP's historical EM for experiment:", e)
    #find the list of models for that experiment
    model = get_models_for_experiment(e)
    
    for m in model:
        #looping through each model
        print('Model:', m)
        
        #creating the filepath to the directory containing the CVDP output historical ensemble mean - one per model
        folder_path_CVDP = '/gws/nopw/j04/extant/users/slbennie/CVDP_outputs/NAO/historical/historical_NAO/'

        #finding CVDP's output historical ensemble mean for this model
        filepath_CVDP = [os.path.join(folder_path_CVDP, filename_CVDP) for filename_CVDP in os.listdir(folder_path_CVDP) if m in filename_CVDP and '_EM' in filename_CVDP][0]
        
        #Opening the file for the ensemble mean and calling function to crop to the NA region and shift the coordinate system
        historical_NAO_EM = CVDP_EM_crop_NA_sector(filepath_CVDP, pattern)
    
        #now setting up the folder path to get the file names for each ensemble's psl anomaly
        folder_path = '/gws/nopw/j04/extant/users/slbennie/psl_anomalies/'+e+'/'+m+'/'
        
        #Getting the list of file names within the models folder (all the anomalies)
        ens_files = [os.path.join(folder_path, filename) for filename in os.listdir(folder_path) if m in filename]
        print('Number of ensemble members:', len(ens_files))
        
        for i in range(0,len(ens_files)):
            #looping through each ensemble member
            #cropping each anomaly to just the NA sector and whichever time
            anomaly = crop_ens_north_atlantic(ens_files[i], 1850,2014)
            
            #creating the output file name for the NAO index for each file
            output_file = ens_files[i].replace('anomaly.nc', 'NAO_index.nc')
            output_file = output_file.replace('extant/users/slbennie/psl_anomalies/', 'extant/users/slbennie/projection_indicies/NAO/')
        
            #calling function to project the anomaly onto a spatial pattern (in this case the historical ensemble mean calculated from CVDP)
            projection = project_onto_spatial_pattern(anomaly, historical_NAO_EM, variable, output_file)
        print('Number of iterations:',i+1)
```
