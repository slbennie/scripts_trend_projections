```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
from scipy.stats import linregress
import os
import pandas as pd
import matplotlib.offsetbox as offsetbox
```

```python
# Dictionary of standard (LEADER) model colours
col_dict = {'ACCESS-ESM1-5':'#7f3c8d','CanESM5':'#11a579','CMCC-CM2-SR5':'#3969ac','FGOALS-g3':'#f2b701','GISS-E2-1-G':'#e73f74','HadGEM3-GC31-LL':'#80ba5a','IPSL-CM6A-LR':'#e68310','MIROC6':'#008695','MPI-ESM1-2-LR':'#cf1c90','NorESM2-LM':'#f97b72'}
exp_col_dict = {"hist-totalO3": "#66c5cc",
	"hist-sol": 	"#f6cf71",
	"hist-volc":	"#f89c74",
	"unlabeled_1":  "#dcb0f2",  # rename if needed    
	"hist-aer": 	"#87c55f",
	"unlabeled_2":  "#9eb9f3",  # rename if needed
	"hist-GHG": 	"#fe88b1",
	"hist-nat": 	"#c9db74",
	"hist-total03":  "#8be0a4",  # rename if needed    
	"historical":   "#b497e7",
	"observations": "#b3b3b3"
}
```

```python
def get_ens_timeseries_model(folder_path, model):
    #function which will loop through folder to find the ensemble files for each model
    model_ens_files = [os.path.join(folder_path, filename) for filename in os.listdir(folder_path) if model in filename and '_EM' not in filename]
    return model_ens_files

def calc_NAO_trend_changing_period_ens(filepath, folderpath, model, startdate, endyear, startyear, trend_period_initial, variable, variable_ens):
    #calculating the trend between a start and end year. Increasing the period of trend calculation by one year each iteration.
    #filepath = path to ensemble mean
    #folderpath = path to folder where all ensemble members are stored
    #model = str model name
    #startdate = which year to start the calculation from
    #endyear = the actual endyear of the timeseries
    #startyear = the actual startyear of the timeseries
    #trend_period_inital = the initial length of the first trend caclulation
    #variable = variable to perform calculation over 'nao_timeseries_djf_em'
    #variable_ens = same as above but for the ensemble members

    #Creating empty arrays
    NAO_trend_slope = np.zeros(endyear-startdate-trend_period_initial+1)
    NAO_trend_intercept = np.zeros(endyear-startdate-trend_period_initial+1)
    
    #opening dataset
    ds = xr.open_dataset(filepath)

    #putting the startyear as 0 and adjusting the rest to fit this
    startdate = (startdate - startyear)
    enddate = (startdate + trend_period_initial)
    endyear = (endyear - startyear +1)
    trend_period = trend_period_initial
    i = 0

    while enddate < endyear:
        #taking slice of dataset to perform the linregress on
        ds_slice = ds[variable].isel(TIME=slice(startdate,enddate)).values
        
        #time as a simple integer array        
        time_numeric = np.arange(trend_period)
    
        #perform the regression using linregress (returns slope, intercept, r_value, p_value, std_err)
        reg_slope, reg_intercept, _, _, _ = linregress(time_numeric, ds_slice)
    
        #store the results within the arrays made earlier
        NAO_trend_slope[i] = reg_slope
        NAO_trend_intercept[i] = reg_intercept

        enddate = enddate + 1
        trend_period = trend_period + 1
        i = i + 1

    #Now similar but for each ensemble member
    #Need to retrieve the file names for each ensemble member
    model_ens_files = get_ens_timeseries_model(folder_path, model)

    enddate = (startdate + trend_period_initial)
    trend_period = trend_period_initial
    i = 0
    
    #create 2D array, (time,ensembles)
    NAO_trend_slope_ens = np.zeros((endyear-startdate-trend_period_initial,len(model_ens_files)))
    NAO_trend_intercept_ens = np.zeros((endyear-startdate-trend_period_initial,len(model_ens_files)))

    j = 0
    for file in model_ens_files:
        #opening the file provided by the list of files
        ds = xr.open_dataset(file)

        enddate = (startdate + trend_period_initial)
        trend_period = trend_period_initial
        i = 0
        
        while enddate < endyear:
            #Taking slice of dataset to perform the linregress on
            ds_slice = ds[variable_ens].isel(TIME=slice(startdate,enddate)).values
                
            # time as a simple integer array        
            time_numeric = np.arange(trend_period)
            
            # Perform the regression using linregress (returns slope, intercept, r_value, p_value, std_err)
            reg_slope, reg_intercept, _, _, _ = linregress(time_numeric, ds_slice)
            
            # Store the results within the arrays made earlier
            NAO_trend_slope_ens[i,j] = reg_slope
            NAO_trend_intercept_ens[i,j] = reg_intercept
        
            enddate = enddate + 1
            trend_period = trend_period + 1
            i = i + 1
       
        j = j+1

    #now find the max and min across the files
    max_slope = np.max(NAO_trend_slope_ens, axis = 1)
    min_slope = np.min(NAO_trend_slope_ens, axis = 1)

    max_intercept = np.max(NAO_trend_intercept_ens, axis = 1)
    min_intercept = np.min(NAO_trend_intercept_ens, axis = 1)


    return NAO_trend_slope, max_slope, min_slope, NAO_trend_slope_ens #could also pass the intercept in here if useful.
    
def calc_NAO_trend_changing_period(filepath, startdate, endyear, startyear, trend_period_initial, variable):
    #same as above function but only for a single timeseries
    NAO_trend_slope = np.zeros(endyear-startdate-trend_period_initial+1)
    NAO_trend_intercept = np.zeros(endyear-startdate-trend_period_initial+1)

    #opening dataset
    ds = xr.open_dataset(filepath)

    startdate = (startdate - startyear)
    enddate = (startdate + trend_period_initial)
    endyear = (endyear - startyear +1)
    trend_period = trend_period_initial
    i = 0

    while enddate < endyear:
        #Taking slice of dataset to perform the linregress on
        ds_slice = ds[variable].isel(TIME=slice(startdate,enddate)).values
        
        # time as a simple integer array        
        time_numeric = np.arange(trend_period)
    
        # Perform the regression using linregress (returns slope, intercept, r_value, p_value, std_err)
        reg_slope, reg_intercept, _, _, _ = linregress(time_numeric, ds_slice)
    
        # Store the results within the arrays made earlier
        NAO_trend_slope[i] = reg_slope
        NAO_trend_intercept[i] = reg_intercept

        #increasing the enddate, trend_period ( to update the numeric time array) and the index i.
        enddate = enddate + 1
        trend_period = trend_period + 1
        i = i + 1

    return NAO_trend_slope #could also ass the intercept in ehre if useful.
```

```python
#plotting multi-model mean for each experiment

def get_ensemble_means_timeseries(folder_path):
    #function which will loop through a folder (my folders are by experiment) to find all the relevant EM files
    ens_means_files = [os.path.join(folder_path, filename) for filename in os.listdir(folder_path) if '_EM.cvdp_data.' in filename]
    return ens_means_files

def get_multimodel_mean(files, variable):
    #this will open all the files given by the list of files
    #then calculate the mean across them (multimodel mean)

    data_arrays = []

    #looping over all files
    for file in files:

        #opening the dataset for one file
        ds = xr.open_dataset(file, engine='netcdf4')

        #appending the timeseries data to the array
        data_arrays.append(ds[variable])

    #calculating the mean
    multimodel_mean = np.mean([da.values for da in data_arrays], axis=0)

    #creating an array of the mean
    multimodel_da = xr.DataArray(multimodel_mean, coords=data_arrays[0].coords, dims=data_arrays[0].dims)

    return multimodel_da

fig,ax = plt.subplots(figsize=(14,5))
mode = 'NAO'
variable = 'nao_timeseries_djf_em'
experiment =['historical','hist-aer','hist-GHG','hist-totalO3','hist-sol','hist-volc']
times = np.arange(1850, 2015)

experiment_mmm = {}

#looping over all experiments to calculate the multimodel mean (mmm) and then the running mean
for e in experiment:
    
    folder_path = '/../../gws/nopw/j04/extant/users/slbennie/CVDP_outputs/'+ mode + '/' + e + '/' + e + '_' + mode

    ens_means_files = get_ensemble_means_timeseries(folder_path)

    multimodel_da = get_multimodel_mean(ens_means_files, variable)

    experiment_mmm[e] = multimodel_da.values

window_size = 10

for e in experiment:
    ax.plot(times, experiment_mmm[e], label=e)

    running_mean_mmm = pd.Series(experiment_mmm[e]).rolling(window=window_size, center=True).mean()

    ax.plot(times, running_mean_mmm, label=f'{e} (running mean)', linestyle='--')
#Now time for ERA5

era5 = xr.open_dataset('/gws/nopw/j04/extant/users/slbennie/CVDP_outputs/NAO/historical/historical_NAO/ERA5_1.cvdp_data.1940-2014.nc')
era5_timeseries = era5['nao_timeseries_djf']
era5_times = np.arange(1940,2015)

ax.plot(era5_times, era5_timeseries.values, color='black')
#print(era5_timeseries.values, era5_times)

ax.set_title('Multimodel Means for the NAO index')
ax.set_xlabel('Time (years)')
ax.set_ylabel('NAO Index')
ax.legend(loc=(1.04, 0))
```
