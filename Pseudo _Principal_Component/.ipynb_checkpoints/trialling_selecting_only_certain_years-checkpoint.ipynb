```python
#can use continue.
#want to add in a bit to select the inital and final years
#going to presume it is always monthly data for now
#also just sticking with calculatingdjf for now.
#year_init and year_final in cftime.DatetimeNoLeap
#period = year_final - year_initial
#take the slice
#if len(slice /= period)
#print error with file, print the start, end dates of file as well as length.
#continue
#I will probably need to add in some GISS ones but will need the ensemble mean from CVDP which is a bit trickier - could run seperatley as I have done and then combine?

def get_time_bounds(calendar_type):
    if calendar_type == cftime.DatetimeNoLeap:
        return cftime.DatetimeNoLeap(1850,1,16), cftime.DatetimeNoLeap(2015,1,16)
    elif calendar_type == cftime.Datetime360Day:
        return cftime.Datetime360Day(1850,1,16), cftime.Datetime360Day(2015,1,16)
    else:
        return datetime(1850,1,16), datetime(2015,1,16)

log_file = 'time_check/historical_time_check.txt'

model = ['ACCESS-ESM1-5','CanESM5','CMCC-CM2-SR5','FGOALS-g3','GISS-E2-1-G','HadGEM3-GC31-LL','IPSL-CM6A-LR','MIROC6','MPI-ESM1-2-LR','NorESM2-LM']#historical

period = (2015-1850)*12

with open(log_file, 'w') as f:
    for m in model:

        #writing model name to log file
        f.write(f"\nModel: {m}\n")

        #assigning folder path
        folder_path = '/gws/nopw/j04/leader_epesc/CMIP6_SinglForcHistSimul/InterpolatedFlds/psl/historical/'+m+'/'
        
        #find files and put into list
        ens_files = [os.path.join(folder_path, filename) for filename in os.listdir(folder_path) if m in filename]
        
        #loop through each file in ens_files
        for i in range(0,len(ens_files)):
            with xr.open_dataset(ens_files[i]) as ds:
                #first noting down the first and last time values as well as the number of values inbetween them
                log_line = f"{os.path.basename(ens_files[i])}: {ds['time'].values[0]} to {ds['time'].values[-1]}, {len(ds['time']} time steps\n"
                f.write(log_line)
                
                calendar_type = type(ds.time.values[0])
                year_init, year_final = get_time_bounds(calendar_type)
                ds_slice = ds.sel(time=slice(year_init, year_final))
                
                if len(ds_slice['psl']) != period:
                    print(ens_files[i], ds['time'][0].values, ds['time'][-1].values, len(ds['psl']))
                    continue
            
```
