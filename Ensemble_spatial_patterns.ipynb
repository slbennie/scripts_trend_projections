```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
plt.rcParams.update({'font.size': 20})
```

```python
NAO_rmEM = xr.open_dataset("../NAO/historical_NAO_rmEM/HadGEM3-GC31-LL_r11i1p1f3.cvdp_data.1850-2014.nc")
NAO_pattern_djf_rmEM = NAO_rmEM['nao_pattern_djf']
NAO_pattern_djf_rmEM.plot()
NAO = xr.open_dataset('../NAO/historical_NAO/HadGEM3-GC31-LL_r11i1p1f3.cvdp_data.1850-2014.nc')
NAO_pattern_djf = NAO['nao_pattern_djf']

NAO_EM = xr.open_dataset("../NAO/historical_NAO/HadGEM3-GC31-LL_EM.cvdp_data.nao.1850-2014.nc")
NAO_pattern_djf_EM = NAO_EM['nao_pattern_djf_em']
NAO_EM_rmEM = xr.open_dataset("../NAO/historical_NAO_rmEM/HadGEM3-GC31-LL_EM.cvdp_data.nao.1850-2014.nc")
NAO_pattern_djf_EM_rmEM = NAO_EM_rmEM['nao_pattern_djf_em']
```

```python
pressure_data = [NAO_pattern_djf_rmEM, NAO_pattern_djf, NAO_pattern_djf_EM, NAO_pattern_djf_EM_rmEM]
vmin = min(pressure.min().values for pressure in pressure_data)
vmax = max(pressure.max().values for pressure in pressure_data)

def ensemble_spatial_pattern_plot(pressure, title, vmin, vmax, ax):
    # Create a figure with Cartopy projection
    pressure_plot = pressure.plot(ax=ax, transform=ccrs.PlateCarree(), cmap='coolwarm', vmin=vmin, vmax=vmax, add_colorbar=False)
    
    # Add continent outlines and other geographical features
    ax.coastlines(resolution='110m', color='black', linewidth=1)
    ax.add_feature(cfeature.LAND, edgecolor='black', facecolor='lightgray')
    
    # Add gridlines for latitude/longitude
    ax.gridlines(draw_labels=True, linewidth=0.5, color='gray', linestyle='--', alpha=0.6)

    # Set the title for each subplot
    ax.set_title(f'{title}', fontsize=24, fontweight='bold')

    return pressure_plot

fig, axes = plt.subplots(2,2, subplot_kw={'projection': ccrs.PlateCarree()}, figsize=(20,10))

ensemble_spatial_pattern_plot(NAO_pattern_djf, '(a) Ensemble Member',vmin,vmax, axes[0,0])
ens_mean = ensemble_spatial_pattern_plot(NAO_pattern_djf_EM, '(b) Ensemble Mean',vmin,vmax,axes[0,1])    
pure_int_var = ensemble_spatial_pattern_plot(NAO_pattern_djf_rmEM, '(c) Pure Internal Variability',vmin,vmax,axes[1,0])

axes[1,1].axis('off')

#Adding a single colourbar for the subplot
#cbar = plt.colorbar(pure_int_var, ax=axes[:, :2], orientation='horizontal', shrink=0.5, pad=0.1)
cbar = fig.colorbar(pure_int_var, ax=axes[1,1], orientation='horizontal', pad=0.1)

cbar.set_label('Surface Pressure (hPa)')

plt.tight_layout(rect=[0,0.05,0.95,0.95])
plt.subplots_adjust(hspace=0.5, wspace=0.1)  # Increase vertical and horizontal spacing

plt.savefig('NAO_plots/spatial_pattern_Hadgem3_historical')
plt.show()
```
